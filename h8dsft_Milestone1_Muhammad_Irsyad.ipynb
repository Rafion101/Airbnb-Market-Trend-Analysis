```python
# Using turkey's rule to remove outliers (Price)

# Finding Quartile 1 (q1)
q1 = df.price.quantile(0.25)

# Finding Quartile 3 (q3)
q3 = df.price.quantile(0.75)

# Calculating the interquartile range
iqr = q3-q1

# Calculating the upper bound and Lower bound
upBound = q3 + (1.5*iqr)
lowBound = q1 - (1.5*iqr)

# Printing out the iqr, lower bound and upper bound
print(f"Value of iqr: {iqr}")
print(f'Upper Boundary: {upBound:.2f}' )
print(f'Lower Boundary: {lowBound:.2f}')
```

```python
outlier = df[(df.price < lowBound) | (df.price > upBound)]
noOutlier = df[(df.price>=lowBound) & (df.price<=upBound)]

print(f'Percentage of outliers: {len(outlier)*100/len(df.price):.2f}%%')
```

```python
df = noOutlier

print(f"The price skewness is : {df['price'].skew()}")
print(f"The price kurtosis is : {df['price'].kurtosis()}")
```

## Kendall Corelation Between Number of Reviews and Review per Month

We would like to find a corelation between the neighborhood_group and price, in order for use to maximize of area to market with the following hypothesis.


- H0 : There is no significant difference price neighborhood_group
- H1 : There is a significant difference price neighborhood_group

```python
corr_tau, pval_k = stats.kendalltau(df["number_of_reviews"], df["reviews_per_month"])

print(f"kendall Corellation: ", {corr_tau})
print(f"P-Value: ", {pval_k})
```

- Reason why we use the kendall test is because the data is not a normally distributed data thus we used this instead of spearman

- Based on the corelation analysis we can see that the p value is 0 and the kendall tau value is less than 0.05, which can indicate that there is indeed a corelation between reviews and reviews per months thus <b>we fail to reject the H0 hypothesis</b>.

-  Based on that the number_of_reviews outlier does not need to be removed because the reviews_per_month has already been removed

## ANOVA INFERANTIAL STATISTICS HYPOTHESIS TEST

We can see from the data above that the entire apartment is usually has the largest price compared to other types of rooms

We would like to find a difference between the neighborhood_group and price, in order for us to maximize marketing within the area with the following hypothesis:


- H0 : There is no significant difference price neighborhood_group
- H1 : There is a significant difference price neighborhood_group

```python
# Creating a variable of price per location
manhattan = df[df["neighbourhood_group"] == "Manhattan"]
brooklyn = df[df["neighbourhood_group"] == "Brooklyn"]
queens = df[df["neighbourhood_group"] == "Queens"]
island = df[df["neighbourhood_group"] == "Staten Island"]
bronx = df[df["neighbourhood_group"] == "Bronx"]

# Price of airbnb in each area
manhattanPrice = manhattan["price"]
brooklynPrice = brooklyn["price"]
queensPrice = queens["price"]
islandPrice = island["price"]
bronxPrice = bronx["price"]
```

Creating a variable to hold prices based on locations

```python
# One Way Anova Test
res = stats.f_oneway(manhattanPrice, brooklynPrice, queensPrice, islandPrice, bronxPrice)

# Print P-Value
print('P-value:',res.pvalue)
```

### Result of the ANOVA Test

- Based on the ANOVA test we which compare prices from each different neighbor groups, we gained a <b>p-value of 0</b>

- we can conclude that there is a significant difference price between neighbor groups.

- based on that statemnt we can say that we <b>reject the H0 hypothesis</b>
